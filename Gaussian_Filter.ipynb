## Gaussian Distribution

The Gaussian distribution is probably the most important distribution in all of probability and statistics.
Many populations have distributions that can be fit very closely by an appropriate normal (or Gaussian, bell) curve.
In $ Eq (1) $ We can represent the 2-D gaussian disruption as:


$$ G(x,y)= \frac{1}{\sqrt{2\pi\sigma}}e^{\frac{x^2+y^2}{2\sigma^2}} \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ ...Eq(1) $$  

## Gaussian Kernal

The Gaussian smoothing operator is a 2-D  convolution operator that is used to blur images and  remove detail and noise,especially the gaussian noise . In this sense it is like the mean filter, but it uses a different kernel that represents the shape of a Gaussian (`bell-shaped') hump.

the main principle behind Gaussian distribution is taking the 2-D distribution as a `point-spread' function, and this is achieved by convolution. Since the image is stored as a collection of discrete pixels, we need to produce a discrete approximation to the Gaussian function before we can perform the convolution. 


Gaussian kernel, as its name implies, has the shape of the function “Gaussian distribution” to define the weights inside the kernel, which are used to compute the weighted average of the neighboring points (pixels) in an image.

## OpenCV

in order to implement our Gaussian filter, we shall use the OpenCV library in python.

OpenCV is a free open source library used in real-time image processing. It’s used to process images, videos, and even live streams

```python
import cv2
```

to see the performance of the Gaussian filter we shall apply it to an Image that's affected by Gaussian noise.



using "`cv2.imread`" to process the inputed image, we then sort the 3 by 3 filter kernal into `np.array([]) `Inline code`, the resultant filtered image will be given by using the function 'filter2D' that we can import from 'cv2' libary, the result will look like : 
 




```python

import cv2
import numpy as np
import matplotlib.pyplot as plt
image = cv2.imread('/content/Hand.jpeg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
kernel = np.array([[1/16,1/8,1/16],
                   [1/8, 1/4, 1/8],
                   [1/16,1/8,1/16]])
img = cv2.filter2D(image, -1, kernel)
fig, ax = plt.subplots(1,2,figsize=(10,6))
ax[0].imshow(image)
ax[1].imshow(img)
plt.show()
```

```python

import cv2
import numpy as np
import matplotlib.pyplot as plt
image = cv2.imread('/content/Hand.jpeg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
kernel = np.array([[1,1,1],
                   [1,1,1],
                   [1,1,1]])
img = cv2.filter2D(image, -1, kernel)
fig, ax = plt.subplots(1,2,figsize=(10,6))
ax[0].imshow(image)
ax[1].imshow(img)
plt.show()
```

```python

import cv2
import numpy as np
import matplotlib.pyplot as plt
image = cv2.imread('/content/Gaussian_high_noise_zoomedin.png')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
kernel = (1/273) * np.array([[1,4,7,4,1],
                   [4,16,26,16,4],
                   [7,26,41,26,7],
                   [4,16,26,16,4],
                   [1,4,7,4,1]])
img = cv2.filter2D(image, -1, kernel)
fig, ax = plt.subplots(1,2,figsize=(10,6))
ax[0].imshow(image)
ax[1].imshow(img)
plt.show()
```
